```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
```

```python
v=dataset['reading score']
```

```python
u=dataset['writing score']
```

```python
plt.figure(figsize=(10,4), dpi=140)

plt.subplot(1,2,1)
plt.scatter(v,u,color="red",edgecolor='black')
plt.xlabel('WRITING SCORE')
plt.ylabel('READING SCORE')
plt.title("COMPARISION BETWEEN THE SCORES")
plt.xlim(0, 100); plt.ylim(0, 100)

plt.subplot(1,2,2)
plt.bar(v,u,align='center',edgecolor='black',color='pink')
plt.xlabel('WRITING SCORE')
plt.ylabel('READING SCORE')
plt.title("COMPARISION BETWEEN THE SCORES")
```

CONCLUSION: Reading score and writing score are lineraly deendent with the increase in writing score reading score increases linearly.

```python
plt.rcParams['figure.figsize']=(15,5)
plt.subplot(1,3,1)
plt.title('Maths score',fontsize=30)
sns.barplot(x='gender',y='math score',data=dataset,color='salmon',saturation=.9,edgecolor='black')
plt.subplot(1,3,2)
plt.title('Reading score',fontsize=30)
sns.barplot(x='gender',y='reading score',data=dataset,color='salmon',saturation=.9,edgecolor='black')
plt.subplot(1,3,3)
plt.title('Writing score',fontsize=30)
sns.barplot(x='gender',y='writing score',data=dataset,color='salmon',saturation=.9,edgecolor='black')
```

```python
plt.rcParams['figure.figsize']=(15,5)
plt.subplot(1,3,1)
plt.title('Maths score',fontsize=30)
sns.barplot(hue='gender',x='test preparation course',y='math score',data=dataset,color='salmon',saturation=.9
           ,edgecolor='black')
plt.subplot(1,3,2)
plt.title('Reading score',fontsize=30)
sns.barplot(hue='gender',x='test preparation course',y='reading score',data=dataset,color='salmon',saturation=.9
           ,edgecolor='black')
plt.subplot(1,3,3)
plt.title('Writing score',fontsize=30)
sns.barplot(hue='gender',x='test preparation course',y='writing score',data=dataset,color='salmon',saturation=.9
           ,edgecolor='black')
```

CONCLUSION: form above it can be concluded that students who have completed thier test preparation coures have scored more in all the three subjects.
